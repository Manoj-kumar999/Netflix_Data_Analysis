```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the datase
df = pd.read_csv("netflix_titles.csv")

# Show the top 5 rows
df.head()
```

Let’s check for missing data first — this helps show you’re good at data cleaning.


```python
df.isnull().sum()
```

## Step 4: Drop unnecessary columns or handle nulls

```python
df.dropna(subset=["director","country"],inplace = True)
df.fillna("Not Specified",inplace=True)
df
```

## Step 5: Count of Movies vs TV Shows

```python
sns.countplot(data =df,x ="type",palette ="pastel")
plt.title("Count of Movies vs TV Shows on Netflex")
plt.xlabel("Type of Content")
plt.ylabel("Count")
plt.show()
```

## Step 6: Top 10 Countries by Number of Shows

```python
top_countries = df["country"].value_counts().head(10)

top_countries.plot(kind="barh",color ="blue")
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Shows")
plt.ylabel("Country")
plt.show()
```

## Step 7: Analyze Top Genres

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("netflix_titles.csv")
# Let's split and count individual genres
df['listed_in'] = df['listed_in'].str.split(', ')
genres = df.explode('listed_in')

# Top 10 genres
top_genres = genres['listed_in'].value_counts().head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='coolwarm')
plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Number of Shows")
plt.ylabel("Genre")
plt.show()
```

## Step 8: Year-wise Content Added

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind = 'bar',figsize = (17,6),color = 'teal')
plt.title("Content Added of Netflex Over the Years")
plt.xlabel("Years")
plt.ylabel("Number of Shows Added")
plt.grid(True)
plt.show()
```

## ALL STEPS  IN ONE CODE OF Netflix_Data_Analysis 

```python
# 1. Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Show plots in notebook
%matplotlib inline

# 2. Load Dataset
df = pd.read_csv("netflix_titles.csv")

# 3. View First Few Rows
df.head()

# 4. Check for Null Values
df.isnull().sum()

# 5. Handle Missing Data
df.dropna(subset=['director', 'country'], inplace=True)
df.fillna("Not Specified", inplace=True)

# 6. Type of Content: Movie vs TV Show
sns.countplot(data=df, x='type', palette='pastel')
plt.title("Count of Movies vs TV Shows on Netflix")
plt.xlabel("Type of Content")
plt.ylabel("Count")
plt.show()

# 7. Top 10 Countries Producing Netflix Content
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='barh', color='skyblue')
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Shows")
plt.ylabel("Country")
plt.show()

# 8. Top 10 Most Popular Genres
df['listed_in'] = df['listed_in'].str.split(', ')
genres = df.explode('listed_in')
top_genres = genres['listed_in'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='coolwarm')
plt.title("Top 10 Most Popular Genres on Netflix")
plt.xlabel("Number of Shows")
plt.ylabel("Genre")
plt.show()

# 9. Content Added Over the Years
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['year_added'] = df['date_added'].dt.year

df['year_added'].value_counts().sort_index().plot(kind='bar', figsize=(12,6), color='teal')
plt.title("Content Added to Netflix Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Shows Added")
plt.grid(True)
plt.show()
```
